```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as pl
```

```python
s=1.0
x=np.linspace(1.03,1.2,100)
Uo=1.0

U=Uo*((s/x)**12-(s/x)**6)

pl.plot(x,U)
pl.grid()
```

```python
x0=2**(1/6)*s
k=9*2**(2/3)*Uo/s**2

Ut=-Uo/4.0 + 0.5*k*(x-x0)**2
pl.plot(x,Ut,'g-')
pl.plot(x,U,'r-')
```
